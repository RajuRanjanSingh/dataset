```python
data1=pd.read_csv('employee_data.csv')
data2=pd.read_csv('manager_data.csv')
df=pd.merge(data1,data2,on='employee_name',how='outer')
```

```python
df=df.drop(['gender'],axis=1)
```

```python
df
```

```python
leave_duration =df['leave_duration'].apply(np.ceil)
 
```

```python
print(leave_duration)
```

```python
for i in range(len(leave_duration)):
    df['leave_duration'][i] = pd.to_datetime(df['leave_dates'][i]) + pd.DateOffset(days= leave_duration[i])
```
